```python
import pandas as pd
import matplotlib.pyplot as plt
```


## **VGG**

```python
v_list = ['CV_5','CV_6','CV_7', 'CV_8', 'CV_9','CV_10','CV_11']
```

```python
v_dictionary = {}
for v in v_list:
  v_dictionary["%s" %v] = pd.read_csv('./'+v+'.csv')
```

```python
v_dictionary['CV_5']['Accuracy']
```

```python
for key in v_dictionary:
  plt.plot(v_dictionary[key]['Epoch'], v_dictionary[key]['Training loss'])

plt.xticks(v_dictionary[key]['Epoch'])
plt.xlabel('Iteration')
plt.ylabel('Training loss')
#plt.legend(loc=1)
plt.grid(True)
plt.tight_layout()
plt.title('Training loss of VGG-16 models with different parameters')
plt.savefig('CIFAR10_VGG_TrainLoss.png', bbox_inches='tight')
plt.show()
```

```python
for key in v_dictionary:
  plt.plot(v_dictionary[key]['Epoch'], v_dictionary[key]['Validation loss'], label=key)

plt.xticks(v_dictionary[key]['Epoch'])
plt.xlabel('Iteration')
plt.ylabel('Validation loss')
#plt.legend(loc=1)
plt.grid(True)
plt.tight_layout()
plt.title('Validation loss of VGG-16 models with different parameters')
plt.savefig('CIFAR10_VGG_ValidationLoss.png', bbox_inches='tight')
plt.show()
```

```python
for key in v_dictionary:
  plt.plot(v_dictionary[key]['Epoch'], v_dictionary[key]['Accuracy'], label=key)

plt.xticks(v_dictionary[key]['Epoch'])
plt.xlabel('Iteration')
plt.ylabel('Accuracy')
#plt.legend(loc=4)
plt.grid(True)
plt.tight_layout()
plt.title('Accuracy of VGG-16 models with different parameters')
plt.savefig('CIFAR10_VGG_Accuracy.png', bbox_inches='tight')
plt.show()
```

## 3 models comparison

```python
all_list = ['VGG16','GoogLeNet','ResNet18']
```

```python
all_dictionary = {}
for a in all_list:
  all_dictionary["%s" %a] = pd.read_csv('./'+a+'.csv')
```

```python
for key in all_dictionary:
  plt.plot(all_dictionary[key]['Epoch'], all_dictionary[key]['Accuracy'], label=key)

plt.xticks([num for num in all_dictionary[key]['Epoch'] if num % 2 == 0])
plt.xlabel('Iteration')
plt.ylabel('Accuracy')
plt.legend(loc=4)
plt.grid(True)
plt.tight_layout()
plt.title('Accuracy of VGG-16, GoogLeNet and ResNet-18 on CIFAR-10')
plt.savefig('CIFAR10_All_Accuracy.png', bbox_inches='tight')
plt.show()
```
